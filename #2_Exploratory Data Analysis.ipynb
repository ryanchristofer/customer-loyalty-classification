```python
# FUNGSI UNTUK VISUALISASI
fig = plt.figure(figsize=(12, 5))

def vis_cat(cat_column, title_count, xy_label):
  ax = fig.add_axes([0,0,1.5,1])
  data1[cat_column].value_counts().plot.pie(autopct='%1.2f%%', shadow=True, figsize=(12, 5))
  plt.xlabel(xy_label)
  plt.ylabel('')

  plt.savefig('fig.png', bbox_inches='tight')
```

```python
# count of gender
df_data1 = data1.groupby('gender_x').agg(count_of_gender=('gender_x','count'))
df_data1
```

```python
# pesebaran data berdasarkan gender
col = 'gender_x'
title_c = 'Distribution of Gender'
label = 'Gender'

vis_cat(col, title_c, label)
```

```python
# Fungsi menampilkan tabel berdasarkan nama kolom
def table(col_name):
    x = data1.groupby(['is_loyal', col_name])['customer_id'].nunique().reset_index(name='Customer')
    x1 = data1.groupby([col_name])['customer_id'].nunique().reset_index(name='AllCustomer')
    x = x.merge(x1, on=col_name)
    x['Percentage'] = round(x['Customer']*100/x['AllCustomer'],2)
    return x
```

```python
col_name = 'gender_x'

table(col_name)
```

```python
col_name = 'class_age'

table(col_name)
```
